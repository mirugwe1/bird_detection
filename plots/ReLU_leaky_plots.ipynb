# The ReLU plot

Rectified Linear Unit,
          
                     f(x)=max(0,x)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

# Rectified Linear Unit (ReLU) Activation Function

# Generating data For Graph
x = np.linspace(-6, 6, 1000)
y = np.maximum(0, x)

# Graph
plt.figure(figsize=(10, 6))
plt.plot(x, y, color="blue")
plt.legend(['Relu'],fontsize=20)
plt.axvline(x=0, color = "black")
plt.xlabel("z",fontsize=20)
plt.ylabel("ReLU(z)",fontsize=20)
plt.savefig('ReLU.pdf')
plt.show()
```

# Leaky ReLU Activation Function

```python
# Leaky Rectified Linear Unit (leaky ReLU) Activation Function
def leaky_ReLU(x):
  data = [max(0.05*value,value) for value in x]
  return np.array(data, dtype=float)


# Generating data For Graph
x_data = np.linspace(-6,6,1000)
y_data = leaky_ReLU(x_data)

# Graph
plt.figure(figsize=(10, 6))
plt.plot(x_data, y_data, color="blue")
plt.axvline(x=0, color = "black")
plt.axhline(y=0, color = "black")
plt.legend(['leaky ReLU'],fontsize=20)
plt.xlabel("z",fontsize=20)
plt.ylabel("f(z)",fontsize=20)
plt.savefig('LEAKY.pdf')
plt.show()
```
